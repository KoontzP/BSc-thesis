```python
import re
import matplotlib.pyplot as plt

file_path = r"Optuna/optunaScriptRunTerminalOutput"

trial_pattern = re.compile(r"Trial (\d+) finished with value: ([-\d.e]+)")

trials = []
values = []

with open(file_path, "r") as file:
    for line in file:
        match = trial_pattern.search(line)
        if match:
            trial_number = int(match.group(1))
            objective_value = float(match.group(2))
            trials.append(trial_number)
            values.append(objective_value)

plt.figure(figsize=(10, 5))
plt.plot(trials, values, marker="o", linestyle="-", color="b", label="Objective Value")
plt.axhline(y=max(values), color="r", linestyle="--", label=f"Best Value: {max(values):.3f}")
plt.xlabel("Trial Number")
plt.ylabel("Objective Value")
plt.title("PPO Optuna Optimization History")
plt.legend()
plt.grid(True)
plt.show()
```
